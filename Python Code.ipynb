```python
import numpy as np
import pandas as pd
import datetime
```

```python
#CPI Excel file
cpi_monthly=pd.read_excel('/content/CPI Data.xlsx',sheet_name='monthly')
cpi_annual=pd.read_excel('/content/CPI Data.xlsx',sheet_name='annual')

#Exchange rate file
exchange_monthly=pd.read_excel('/content/Exchange rate.xlsx',sheet_name='monthly')
exchange_annual=pd.read_excel('/content/Exchange rate.xlsx',sheet_name='annual')

#Exports Merchandise file
exports_monthly=pd.read_excel('/content/Exports Merchandise.xlsx',sheet_name='monthly')
exports_quarterly=pd.read_excel('/content/Exports Merchandise.xlsx',sheet_name='quarterly')
exports_annual=pd.read_excel('/content/Exports Merchandise.xlsx',sheet_name='annual')
```

```python
#Filling the nulll values with zeros

cpi_monthly=cpi_monthly.fillna(0)
cpi_annual=cpi_annual.fillna(0)

exports_monthly=exports_monthly.fillna(0)
exports_quarterly=exports_quarterly.fillna(0)
exports_annual=exports_annual.fillna(0)

```

```python
exports_annual=exports_annual.fillna(0)
exports_annual.isna().sum()
cpi_monthly.isna().sum()
```

```python
cpi_monthly.rename(columns={'Unnamed: 0':'year'},inplace=True)


years=[]
for i in range(len(cpi_monthly)):
  years.append(cpi_monthly['year'][i][0:4])

months=[]

for i in range(len(cpi_monthly)):
  months.append(cpi_monthly['year'][i][5:7])

year2=[]
for i in range(len(years)):
  year2.append(years[i]+'-'+months[i])


year3=[]
for i in range(len(year2)):
  year3.append(datetime.datetime.strptime(year2[i], '%Y-%m').date())


cpi_monthly['year']=year3
```

```python
exchange_monthly.rename(columns={'Period':'year'},inplace=True)


years=[]
for i in range(len(exchange_monthly)):
  years.append(exchange_monthly['year'][i][0:4])

months=[]

for i in range(len(exchange_monthly)):
  months.append(exchange_monthly['year'][i][5:7])

year2=[]
for i in range(len(years)):
  year2.append(years[i]+'-'+months[i])


year3=[]
for i in range(len(year2)):
  year3.append(datetime.datetime.strptime(year2[i], '%Y-%m').date())


exchange_monthly['year']=year3
```

```python
exports_monthly.rename(columns={'Unnamed: 0':'year'},inplace=True)


years=[]
for i in range(len(exports_monthly)):
  years.append(exports_monthly['year'][i][0:4])

months=[]

for i in range(len(exports_monthly)):
  months.append(exports_monthly['year'][i][5:7])

year2=[]
for i in range(len(years)):
  year2.append(years[i]+'-'+months[i])


year3=[]
for i in range(len(year2)):
  year3.append(datetime.datetime.strptime(year2[i], '%Y-%m').date())


exports_monthly['year']=year3
```
